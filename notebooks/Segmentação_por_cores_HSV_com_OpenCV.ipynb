```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
```

```python


img = cv2.imread('grama.jpg')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
figsize = (10,10)
plt.figure(figsize=figsize)
plt.imshow(img_rgb)
plt.title('Imagem Original')
plt.show()
```

```python
rgb_green = np.uint8([[[0,255,0]]]) #cvtColor espera um array 3D
hsv_green = cv2.cvtColor(rgb_green, cv2.COLOR_RGB2HSV)[0, 0, :]
print(hsv_green)
```

```python
hsv_img = cv2.cvtColor(img_rgb, cv2.COLOR_RGB2HSV)
l_th = hsv_green - np.array([70, 200, 200])
u_th = hsv_green + np.array([30, 0, 0])
mask = cv2.inRange(hsv_img, l_th, u_th)
plt.figure(figsize=figsize)
plt.imshow(mask)
plt.title('Máscara resultante')
plt.show()
```

```python
rgb_res = cv2.bitwise_and(img_rgb, img_rgb, mask=mask)
plt.figure(figsize=figsize)
plt.imshow(rgb_res)
plt.title('Imagem com a máscara aplicada')
plt.show()
```
